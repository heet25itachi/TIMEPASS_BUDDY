Affine Term Structure Model (ATSM)An affine term structure model is a financial model that relates zero-coupon bond prices (the discount curve) to a spot rate model. It is particularly useful for deriving the yield curve—the process of determining spot rate model inputs from observable bond market data.The affine class of term structure models implies the convenient form that log bond prices are linear functions of the spot rate (and potentially additional state variables).The Basic StructureWe start with a stochastic short rate model with dynamics:$$dr_t = \mu(r_t, t) dt + \sigma(r_t, t) dW_t$$and a risk-free zero-coupon bond maturing at time $T$ with price $P(r_t, t)$ at time $t$.The price of a zero-coupon bond is given by the risk-neutral expectation:$$P(r_t, t) = \mathbb{E}^{\mathbb{Q}}\left[ \exp \left( - \int_t^T r_u du \right) \mid r_t \right]$$where $\mathbb{E}^{\mathbb{Q}}$ is the expectation taken with respect to the risk-neutral probability measure $\mathbb{Q}$.If the bond's price has the form:$$P(r_t, t) = \exp \left( A(T-t) - B(T-t) r_t \right)$$where $A(\tau)$ and $B(\tau)$ are deterministic functions of the time to maturity $\tau = T-t$, then the short rate model is said to have an affine term structure.The yield of a bond with maturity $\tau = T-t$, denoted by $y_t(\tau)$, is given by:$$y_t(\tau) = - \frac{\ln P(r_t, t)}{\tau} = \frac{B(\tau)}{\tau} r_t - \frac{A(\tau)}{\tau}$$This shows that the yield $y_t(\tau)$ is an affine function (linear plus a constant) of the short rate $r_t$.Constraints and Differential EquationsAssuming the bond has an affine term structure, the functions $A(\tau)$ and $B(\tau)$ must satisfy the constraints imposed by the Term Structure Equation (a special case of the Feynman-Kac formula).If we assume the drift $\mu$ and volatility $\sigma^2$ under the risk-neutral measure are affine in the state variable $r_t$:$$\mu(r_t, t) = \alpha + \beta r_t$$$$\sigma^2(r_t, t) = \gamma + \delta r_t$$(where $\alpha, \beta, \gamma, \delta$ are parameters, potentially time-dependent), the differential equations for $A(\tau)$ and $B(\tau)$ become a pair of ordinary differential equations (ODEs).ODE for $B(\tau)$:$$\frac{dB(\tau)}{d\tau} = \beta B(\tau) - \frac{1}{2}\delta B(\tau)^2 - 1$$ODE for $A(\tau)$:$$\frac{dA(\tau)}{d\tau} = \alpha B(\tau) - \frac{1}{2}\gamma B(\tau)^2$$Both $A(\tau)$ and $B(\tau)$ must satisfy the boundary condition:$$A(0) = 0, \quad B(0) = 0$$Examples of Affine Term Structure Models1. Affine Term Structure Model (ATSM)An affine term structure model is a financial model that relates zero-coupon bond prices (the discount curve) to a spot rate model. It is particularly useful for deriving the yield curve—the process of determining spot rate model inputs from observable bond market data.The affine class of term structure models implies the convenient form that log bond prices are linear functions of the spot rate (and potentially additional state variables).The Basic StructureWe start with a stochastic short rate model with dynamics:$$dr_t = \mu(r_t, t) dt + \sigma(r_t, t) dW_t$$and a risk-free zero-coupon bond maturing at time $T$ with price $P(r_t, t)$ at time $t$.The price of a zero-coupon bond is given by the risk-neutral expectation:$$P(r_t, t) = \mathbb{E}^{\mathbb{Q}}\left[ \exp \left( - \int_t^T r_u du \right) \mid r_t \right]$$where $\mathbb{E}^{\mathbb{Q}}$ is the expectation taken with respect to the risk-neutral probability measure $\mathbb{Q}$.If the bond's price has the form:$$P(r_t, t) = \exp \left( A(T-t) - B(T-t) r_t \right)$$where $A(\tau)$ and $B(\tau)$ are deterministic functions of the time to maturity $\tau = T-t$, then the short rate model is said to have an affine term structure.The yield of a bond with maturity $\tau = T-t$, denoted by $y_t(\tau)$, is given by:$$y_t(\tau) = - \frac{\ln P(r_t, t)}{\tau} = \frac{B(\tau)}{\tau} r_t - \frac{A(\tau)}{\tau}$$This shows that the yield $y_t(\tau)$ is an affine function (linear plus a constant) of the short rate $r_t$.Derivation: The Feynman-Kac PDEThe bond price's definition implies a link to the Feynman-Kac formula, which states that the bond's price $P(\tau, x)$ may be explicitly modeled by a Partial Differential Equation (PDE). Assuming the bond price is a function of $x \in \mathbb{R}^n$ latent factors, the price $P$ must satisfy:$$-\frac{\partial P}{\partial \tau} + \sum_{i=1}^n \mu_i \frac{\partial P}{\partial x_i} + \frac{1}{2}\sum_{i,j=1}^n \Omega_{ij} \frac{\partial^2 P}{\partial x_i \partial x_j} - r P = 0$$with the boundary condition $P(0, x)=1$.Here, $\Omega = \Sigma \Sigma^T$ is the covariance matrix of the latent factors, which are assumed to be driven by an Itô Stochastic Differential Equation (SDE) under the risk-neutral measure $\mathbb{Q}$:$$dx = \mu^{\mathbb{Q}} dt + \Sigma dW^{\mathbb{Q}}$$To solve this PDE, we assume an exponential-affine solution for the bond price:$$P(\tau, x) = \exp \left[ A(\tau) + x^T B(\tau) \right], \quad A(0) = B_i(0) = 0$$The partial derivatives required for substitution into the PDE are:$$\frac{\partial P}{\partial \tau} = \left[ A'(\tau) + x^T B'(\tau) \right] P$$$$\frac{\partial P}{\partial x_i} = B_i(\tau) P$$$$\frac{\partial^2 P}{\partial x_i \partial x_j} = B_i(\tau) B_j(\tau) P$$Substituting these derivatives into the PDE and dividing the entire equation by $P$ reduces the partial differential equation to a set of Ordinary Differential Equations (ODEs):$$-\left[ A'(\tau) + x^T B'(\tau) \right] + \sum_{i=1}^n \mu_i B_i(\tau) + \frac{1}{2}\sum_{i,j=1}^n \Omega_{ij} B_i(\tau) B_j(\tau) - r = 0, \quad A(0) = B_i(0) = 0$$Constraints and Affine ODEsTo ensure a closed-form solution (and thus an Affine Term Structure), the instantaneous short rate $r$ and the risk-neutral drift $\mu^{\mathbb{Q}}$ and volatility $\Omega$ of the state factors $x$ must be affine functions of $x$.For the simple single-factor case where $x_t = r_t$:If we assume the drift $\mu$ and volatility $\sigma^2$ under the risk-neutral measure are affine in the state variable $r_t$:$$\mu(r_t, t) = \alpha + \beta r_t$$$$\sigma^2(r_t, t) = \gamma + \delta r_t$$(where $\alpha, \beta, \gamma, \delta$ are parameters, potentially time-dependent), the differential equations for $A(\tau)$ and $B(\tau)$ become a pair of ODEs.ODE for $B(\tau)$:$$\frac{dB(\tau)}{d\tau} = \beta B(\tau) - \frac{1}{2}\delta B(\tau)^2 - 1$$ODE for $A(\tau)$:$$\frac{dA(\tau)}{d\tau} = \alpha B(\tau) - \frac{1}{2}\gamma B(\tau)^2$$Both $A(\tau)$ and $B(\tau)$ must satisfy the boundary condition:$$A(0) = 0, \quad B(0) = 0$$Examples of Affine Term Structure Models1.
```
import numpy as np
from scipy.linalg import expm, inv
from scipy.optimize import least_squares

# --- GLOBAL ECONOMIC DATA (Based on User Prompt) ---
REPO_RATE_RBI = 0.065
INFLATION_TARGET = 0.042
GDP_GROWTH = 0.067
# Mock yield data for calibration (Tenor in years, Yield in decimal)
MOCK_YIELD_DATA = {
    'tenors': np.array([0.5, 1.0, 2.0, 3.0, 5.0, 7.0, 10.0]),
    'yields': np.array([0.063, 0.065, 0.067, 0.069, 0.071, 0.073, 0.075])
}

class AffineTermStructureModel:
    """
    Two-Factor Gaussian Affine Term Structure Model (ATSM) (Level and Slope).
    The factors follow dX_t = kappa_phys * (theta_phys - X_t) dt + sigma dW_t.
    """
    def __init__(self, kappa_phys, theta_phys, sigma, lambda_risk, delta):
        # Physical Parameters (P-measure)
        self.kappa_phys = kappa_phys  # N x N matrix of mean reversion speed
        self.theta_phys = theta_phys  # N x 1 vector of long-run means
        self.sigma = sigma            # N x N matrix of instantaneous volatility

        # Risk-Neutral Parameters (Q-measure)
        self.lambda_risk = lambda_risk # N x N matrix of factor risk price (lambda_0 + lambda_1 * X_t is simplified)
        self.delta = delta             # N x 1 drift adjustment: kappa_Q * theta_Q - kappa_P * theta_P

        # Derived Q-measure parameters
        self.kappa_Q = self.kappa_phys + self.lambda_risk
        self.theta_Q = inv(self.kappa_Q) @ (self.kappa_phys @ self.theta_phys - self.delta)

        # Volatility covariance (Omega = Sigma * Sigma.T)
        self.omega = self.sigma @ self.sigma.T

        # Time grid for integration (used for pre-computation)
        self.tau_max = MOCK_YIELD_DATA['tenors'][-1] + 1.0 # Extend slightly beyond max tenor
        self.tau_grid = np.linspace(0.001, self.tau_max, 500)
        self.dt = self.tau_grid[1] - self.tau_grid[0]

        # Pre-compute affine functions A(tau) and B(tau)
        self._precompute_affine_functions()

        # Initial state guess (used as initial factor value X_0)
        self.X0 = np.array([[REPO_RATE_RBI], [0.0]]) # Level (Repo Rate) and Slope (0)

        print("ATSM Model Initialized: Two-Factor Gaussian (Level-Slope).")


    def _precompute_affine_functions(self):
        """
        Pre-computes the A(tau) scalar and B(tau) vector for all time steps (tau)
        using the matrix exponential solution.
        """
        N = self.kappa_Q.shape[0]

        # 1. Compute B(tau) vector
        self.B_tau_values = np.zeros((N, len(self.tau_grid)))

        # Calculate B(tau) = inv(kappa_Q) * (I - exp(-kappa_Q * tau)) * 1
        # In a multi-factor model, this is typically B(tau)^T = 1.T * inv(kappa_Q) * (I - exp(-kappa_Q * tau))

        # We use the direct solution B(tau) = inv(kappa_Q) * (I - exp(-kappa_Q * tau))
        # Note: B(tau) in the affine formula is usually a column vector (N x 1)

        for i, tau in enumerate(self.tau_grid):
            # Matrix exponential for B(tau)
            exp_K_tau = expm(-self.kappa_Q * tau)
            B_tau = inv(self.kappa_Q) @ (np.eye(N) - exp_K_tau)
            # Take the sum of the rows to get the N x 1 vector for the factor model.
            # B_tau is defined such that the bond price is exp(A - B.T * X).
            # The structure for two-factor Gaussian model typically uses B = sum(B_i) where B_i are components.
            # Here, we follow the common setup where B_tau is the vector (N x 1)
            self.B_tau_values[:, i] = B_tau.sum(axis=1)

        # 2. Compute A(tau) scalar (Integral term)
        self.A_tau_values = np.zeros(len(self.tau_grid))

        # A(tau) = integral from 0 to tau [ B(s).T * kappa_Q * theta_Q - 0.5 * B(s).T * Omega * B(s) ] ds
        # Note: B(s) is the N x 1 vector derived above.

        # The integrand components calculated at each time step j:
        term_B_Q_theta_Q = np.zeros(len(self.tau_grid))
        term_B_Omega_B = np.zeros(len(self.tau_grid))

        for j, tau in enumerate(self.tau_grid):
            B_tau = self.B_tau_values[:, j].reshape(N, 1) # N x 1 vector

            # Term 1: B(s).T * kappa_Q * theta_Q (1 x N * N x N * N x 1 = scalar)
            # This is complex, let's simplify by using the known solution form:
            # integral B(s).T * (kappa_Q * theta_Q) ds = A(tau) component from B(tau)

            # Simplified Gaussian ATSM solution for A(tau)

            # Using the simplified integral for the specific Gaussian case:
            term_B_Q_theta_Q[j] = B_tau.T @ self.kappa_Q @ self.theta_Q

            # Term 2: 0.5 * B(s).T * Omega * B(s) (1 x N * N x N * N x 1 = scalar)
            term_B_Omega_B[j] = 0.5 * B_tau.T @ self.omega @ B_tau

        # The error was in this line, trying to sum a 1D array along an out-of-bounds axis.
        # We assume term_B_Q_theta_Q and term_B_Omega_B are already 1D arrays of scalar values.
        integrand = term_B_Q_theta_Q + term_B_Omega_B # CORRECTED: Removed .sum(axis=1)

        # Integrate over tau_grid using the trapezoidal rule (Cumulative Sum)
        # Note: The integral is cumulative to get A(tau)
        self.A_tau_values = np.cumsum(integrand) * self.dt


    def get_bond_price(self, tau, X_t):
        """
        Calculates the bond price P(t, T) = exp(A(tau) - B(tau).T * X_t)
        where tau = T - t.
        """
        if tau <= 0.001:
            return 1.0 # P(t, t) = 1

        # Find the index for the closest tau in the pre-computed grid
        idx = np.argmin(np.abs(self.tau_grid - tau))

        A_tau = self.A_tau_values[idx]
        B_tau = self.B_tau_values[:, idx].reshape(len(self.kappa_Q), 1)

        # Bond Price Formula: P(t, T) = exp(A(tau) - B(tau).T * X_t)
        return np.exp(A_tau - B_tau.T @ X_t)[0, 0]


    def get_yield(self, tau, X_t):
        """
        Calculates the zero-coupon yield Y(t, T) = -log(P(t, T)) / tau
        """
        price = self.get_bond_price(tau, X_t)
        if tau <= 0.001:
            return X_t[0, 0] # Short rate is the instantaneous yield
        return -np.log(price) / tau


class EconomicScenarioEngine:
    """Manages parameter adjustments for different economic scenarios."""
    def __init__(self, N_factors=2):
        self.N = N_factors
        # Baseline Parameters (General structure based on research literature)
        self.base_params = {
            'kappa_phys': np.array([[0.1, 0.0], [0.0, 0.5]]),  # Factors mean-revert independently
            'theta_phys': np.array([[0.05], [0.0]]),            # Long-run Level at 5%, Slope at 0
            'sigma': np.array([[0.005, 0.0], [0.0, 0.01]]),    # Low volatility on Level, higher on Slope
            'lambda_risk': np.array([[0.1, 0.0], [0.0, 0.2]]), # Risk prices
            'delta': np.array([[0.0], [0.0]])                  # Initial drift adjustment
        }

    def adjust_model_params(self, name):
        """Adjusts the baseline parameters for a specific named scenario."""
        params = self.base_params.copy()

        # Use the current RBI Repo Rate as a key anchor for the Level factor (X1)
        repo_rate_target = REPO_RATE_RBI

        # Scenario adjustments based on RBI risks (mocked adjustments)
        if name == 'Baseline':
            # Adjust theta_phys[0] (Level) to anchor around the expected long-term rate
            params['theta_phys'][0] = [repo_rate_target - 0.005] # Target just below current repo

        elif name == 'High_Inflation':
            # Higher long-run mean (theta) due to persistent inflation
            params['theta_phys'][0] = [repo_rate_target + 0.008]
            params['sigma'][0, 0] = 0.007 # Higher volatility
            params['lambda_risk'][0, 0] = 0.2 # Higher level risk price

        elif name == 'Slow_Growth':
            # Lower long-run mean (theta) due to recessionary pressure
            params['theta_phys'][0] = [repo_rate_target - 0.015]
            params['sigma'][1, 1] = 0.008 # Lower slope volatility (less curve steepening)

        elif name == 'Global_Taper':
            # Increased long-run steepness (theta[1]) and higher volatility
            params['theta_phys'][1] = [0.003] # Positive long-run slope
            params['sigma'][0, 0] = 0.01
            params['sigma'][1, 1] = 0.02

        else:
            raise ValueError(f"Unknown scenario name: {name}")

        # Recalculate delta to maintain X0 as the starting point.
        # X0 (Level) = Repo Rate.
        # If X0 is an equilibrium point (X0 = theta_phys), then delta = 0.
        # Since X0 is *not* necessarily theta_phys, we ensure the short rate
        # at time 0 (r_0 = X0[0]) matches the starting short rate (Repo rate).

        # For simplicity in this fix, we will set delta to be zero and assume
        # the calibration step (least_squares) will move X0 away from theta_phys
        # if necessary, but this allows the model to initialize correctly.
        params['delta'][0] = repo_rate_target - params['theta_phys'][0]

        return AffineTermStructureModel(**params)


def objective_function(X0_params, model, tenors, actual_yields):
    """
    Objective function for Least Squares: Minimize the difference between
    model-implied yields and actual market yields.

    X0_params is the vector of initial factor values [X1_0, X2_0, ...]
    """
    X0 = X0_params.reshape(-1, 1)

    # Calculate model-implied yields
    model_yields = [model.get_yield(tau, X0) for tau in tenors]
    model_yields = np.array(model_yields)

    # Return the difference vector (residuals)
    return model_yields - actual_yields


def calibrate_and_simulate(scenarios_engine, yield_data):
    """
    Initializes and calibrates the baseline model, then simulates all scenarios.
    """
    tenors = yield_data['tenors']
    actual_yields = yield_data['yields']

    # 1. Initialize and Calibrate Baseline Model
    print("\n--- Calibration (Baseline Scenario) ---")
    base_model = scenarios_engine.adjust_model_params('Baseline')

    # Use long-run mean as initial guess for X0 for calibration
    x0_guess = base_model.theta_phys.flatten()

    # Perform Least Squares optimization to find X0 that minimizes yield error
    res = least_squares(
        objective_function,
        x0_guess,
        args=(base_model, tenors, actual_yields),
        method='lm',
        verbose=0 # Hide optimization output
    )

    X0_calibrated = res.x.reshape(-1, 1)
    base_model.X0 = X0_calibrated # Update the model with calibrated X0

    print(f"Calibration successful: X0_calibrated = {X0_calibrated.flatten()}")
    print(f"Initial Yield Error (RMSE): {np.sqrt(np.mean(res.fun**2)) * 100:.3f} bps")

    # 2. Simulate and Compare Scenarios
    results = {}
    scenarios = ['Baseline', 'High_Inflation', 'Slow_Growth', 'Global_Taper']

    for name in scenarios:
        # Re-initialize the model with scenario parameters
        scenario_model = scenarios_engine.adjust_model_params(name)

        # Set the initial factor state to the calibrated baseline X0
        scenario_model.X0 = X0_calibrated

        # Simulate yield curve using the new model parameters (at t=0, X_t=X0)
        scenario_yields = [scenario_model.get_yield(tau, scenario_model.X0) for tau in tenors]

        results[name] = scenario_yields
        print(f"Scenario '{name}' loaded with yields for {tenors.max()} years.")

    return results, base_model

# =============================================================================
# MAIN EXECUTION
# =============================================================================

# Global variables (Mocking the canvas environment variables)
__app_id = 'affine_model_app'

print(f"Initializing Affine Term Structure Model for Indian Economy (2025-10-30)")
print(f"RBI Repo Rate: {REPO_RATE_RBI * 100}%, Inflation Target: {INFLATION_TARGET * 100}%, GDP Growth: {GDP_GROWTH * 100}%")
print("-" * 75)

try:
    econ_scen = EconomicScenarioEngine()

    # Run calibration and simulation
    results, base_model = calibrate_and_simulate(econ_scen, MOCK_YIELD_DATA)

    # --- Display Results ---
    print("\n--- Scenario Analysis Summary ---")
    print(f"{'Tenor (Y)':<10} | {'Baseline':<12} | {'High Infl':<12} | {'Slow Growth':<12} | {'Global Taper':<12}")
    print("-" * 75)

    for i, t in enumerate(MOCK_YIELD_DATA['tenors']):
        baseline_yield = results['Baseline'][i] * 100
        high_infl_yield = results['High_Inflation'][i] * 100
        slow_growth_yield = results['Slow_Growth'][i] * 100
        global_taper_yield = results['Global_Taper'][i] * 100

        print(f"{t:<10.1f} | {baseline_yield:<12.3f} | {high_infl_yield:<12.3f} | {slow_growth_yield:<12.3f} | {global_taper_yield:<12.3f}")

except Exception as e:
    print(f"\nAn error occurred during execution: {e}")
```

```
import numpy as np
from scipy.linalg import expm, inv
from scipy.optimize import least_squares
import matplotlib.pyplot as plt

# --- GLOBAL ECONOMIC DATA (Based on User Prompt) ---
REPO_RATE_RBI = 0.065   # The current short-term policy rate (6.5%)
INFLATION_TARGET = 0.042
GDP_GROWTH = 0.067
# Mock yield data for calibration (Tenor in years, Yield in decimal)
MOCK_YIELD_DATA = {
    'tenors': np.array([0.5, 1.0, 2.0, 3.0, 5.0, 7.0, 10.0]),
    'yields': np.array([0.063, 0.065, 0.067, 0.069, 0.071, 0.073, 0.075]) # Calibrating to a slightly upward-sloping curve
}

# =============================================================================
# CORE ATSM CLASS (TWO-FACTOR GAUSSIAN)
# =============================================================================

class AffineTermStructureModel:
    """
    Two-Factor Gaussian Affine Term Structure Model (Level and Slope).
    The factors (X_t) follow an Ornstein-Uhlenbeck process under the physical (P) measure:
    dX_t = kappa_phys * (theta_phys - X_t) dt + sigma dW_t
    """
    def __init__(self, kappa_phys, theta_phys, sigma, lambda_risk, delta):
        # 1. Physical Parameters (P-measure)
        self.kappa_phys = kappa_phys    # N x N matrix: Speed of mean reversion
        self.theta_phys = theta_phys    # N x 1 vector: Long-run mean of factors
        self.sigma = sigma              # N x N matrix: Instantaneous volatility
        self.omega = self.sigma @ self.sigma.T # Volatility covariance matrix (Sigma * Sigma.T)

        # 2. Risk-Neutral Parameters (Q-measure)
        self.lambda_risk = lambda_risk  # N x N matrix: Market price of factor risk
        self.delta = delta              # N x 1 vector: Drift adjustment (Delta = kappa_P * theta_P - kappa_Q * theta_Q)

        # Derived Q-measure parameters (Key for no-arbitrage pricing)
        # kappa_Q = kappa_P + lambda_risk
        self.kappa_Q = self.kappa_phys + self.lambda_risk

        # theta_Q = inv(kappa_Q) * (kappa_P * theta_P - Delta)
        self.theta_Q = inv(self.kappa_Q) @ (self.kappa_phys @ self.theta_phys - self.delta)

        # Time grid for numerical integration of A(tau)
        self.tau_max = MOCK_YIELD_DATA['tenors'][-1] + 1.0
        self.tau_grid = np.linspace(0.001, self.tau_max, 500)
        self.dt = self.tau_grid[1] - self.tau_grid[0] # Step size for trapezoidal rule

        # Pre-compute the key affine functions A(tau) and B(tau)
        self._precompute_affine_functions()

        # Initial state X0: Calibrated to match the market short-rate (Repo Rate)
        self.X0 = np.array([[REPO_RATE_RBI], [0.0]])

        print("ATSM Model Initialized: Two-Factor Gaussian (Level-Slope) with Matrix Exponential solution.")


    def _precompute_affine_functions(self):
        """
        Calculates the B(tau) vector and A(tau) scalar for all time steps (tau)
        by numerically solving the ODEs/using matrix exponential.

        Bond Price P(t, T) = exp(A(tau) - B(tau).T * X_t) where tau = T-t.
        """
        N = self.kappa_Q.shape[0]

        # 1. Compute B(tau) vector (N x 1 vector for each time step)
        self.B_tau_values = np.zeros((N, len(self.tau_grid)))

        # B(tau) = inv(kappa_Q) * (I - exp(-kappa_Q * tau)) * 1
        # Here, the sum(axis=1) is applied to match the factor structure in the exponent B.T * X
        for i, tau in enumerate(self.tau_grid):
            exp_K_tau = expm(-self.kappa_Q * tau)
            B_tau_matrix = inv(self.kappa_Q) @ (np.eye(N) - exp_K_tau)
            # Sum rows to get the N x 1 factor loading vector.
            self.B_tau_values[:, i] = B_tau_matrix.sum(axis=1)

        # 2. Compute A(tau) scalar (Integral term)
        self.A_tau_values = np.zeros(len(self.tau_grid))

        # Integrand components for A(tau) = Integral [ B(s).T * kappa_Q * theta_Q - 0.5 * B(s).T * Omega * B(s) ] ds
        term_B_Q_theta_Q = np.zeros(len(self.tau_grid)) # Term related to Q-measure mean
        term_B_Omega_B = np.zeros(len(self.tau_grid))     # Term related to volatility

        for j, tau in enumerate(self.tau_grid):
            # Extract B(tau) for this time step and reshape to N x 1 column vector
            B_tau = self.B_tau_values[:, j].reshape(N, 1)

            # Term 1: B(s).T @ kappa_Q @ theta_Q (1xN * NxN * Nx1 = scalar)
            term_B_Q_theta_Q[j] = B_tau.T @ self.kappa_Q @ self.theta_Q

            # Term 2: 0.5 * B(s).T @ Omega @ B(s) (1xN * NxN * Nx1 = scalar)
            term_B_Omega_B[j] = 0.5 * B_tau.T @ self.omega @ B_tau

        # The integrand is the sum of the two 1D arrays (scalar values at each step)
        integrand = term_B_Q_theta_Q + term_B_Omega_B

        # Numerical Integration using Cumulative Sum (Trapezoidal Rule Approximation)
        self.A_tau_values = np.cumsum(integrand) * self.dt


    def get_bond_price(self, tau, X_t):
        """
        Calculates the zero-coupon bond price P(t, T) = exp(A(tau) - B(tau).T * X_t).
        """
        if tau <= 0.001:
            return 1.0

        # Find the index for the closest tau in the pre-computed grid
        idx = np.argmin(np.abs(self.tau_grid - tau))

        A_tau = self.A_tau_values[idx]
        B_tau = self.B_tau_values[:, idx].reshape(len(self.kappa_Q), 1)

        # Bond Price Formula: P(t, T) = exp(A(tau) - B(tau).T * X_t)
        return np.exp(A_tau - B_tau.T @ X_t)[0, 0]


    def get_yield(self, tau, X_t):
        """
        Calculates the zero-coupon yield Y(t, T) = -log(P(t, T)) / tau.
        """
        price = self.get_bond_price(tau, X_t)
        if tau <= 0.001:
            # For tau approaching zero, the yield approximates the short rate (X_t[0,0])
            return X_t[0, 0]
        return -np.log(price) / tau

# =============================================================================
# SCENARIO ENGINE & CALIBRATION/PLOTTING FUNCTIONS
# =============================================================================

class EconomicScenarioEngine:
    """Manages parameter adjustments for different economic scenarios."""
    def __init__(self, N_factors=2):
        self.N = N_factors
        # Baseline Parameters: Chosen to roughly reflect empirical observations for
        # a Level-Slope model (kappa_phys[0,0] is slow, kappa_phys[1,1] is faster).
        self.base_params = {
            'kappa_phys': np.array([[0.1, 0.0], [0.0, 0.5]]),
            'theta_phys': np.array([[0.05], [0.0]]),
            'sigma': np.array([[0.005, 0.0], [0.0, 0.01]]),
            'lambda_risk': np.array([[0.1, 0.0], [0.0, 0.2]]),
            'delta': np.array([[0.0], [0.0]])
        }

    def adjust_model_params(self, name):
        """Adjusts parameters based on economic outlook for yield curve shift analysis."""
        params = self.base_params.copy()
        repo_rate_target = REPO_RATE_RBI

        if name == 'Baseline':
            # Target the long-run mean (theta_phys[0]) slightly below the current repo rate
            params['theta_phys'][0] = [repo_rate_target - 0.005]

        elif name == 'High_Inflation':
            # Higher long-run level (theta_phys[0]) and increased volatility/risk price (lambda_risk)
            params['theta_phys'][0] = [repo_rate_target + 0.008]
            params['sigma'][0, 0] = 0.007
            params['lambda_risk'][0, 0] = 0.2

        elif name == 'Slow_Growth':
            # Lower long-run level (theta_phys[0]) due to potential rate cuts, flatter curve
            params['theta_phys'][0] = [repo_rate_target - 0.015]
            params['sigma'][1, 1] = 0.008 # Lower slope volatility (less steepening risk)

        elif name == 'Global_Taper':
            # Increased long-run steepness (theta_phys[1]) reflecting higher term premium
            # and higher overall volatility due to global uncertainty.
            params['theta_phys'][1] = [0.003]
            params['sigma'][0, 0] = 0.01
            params['sigma'][1, 1] = 0.02

        else:
            raise ValueError(f"Unknown scenario name: {name}")

        # The 'delta' parameter is calculated to ensure the Q-measure mean (theta_Q)
        # is correctly derived from the P-measure parameters.
        params['delta'][0] = repo_rate_target - params['theta_phys'][0]

        return AffineTermStructureModel(**params)


def objective_function(X0_params, model, tenors, actual_yields):
    """Residual function for Least Squares: model_yields - actual_yields."""
    X0 = X0_params.reshape(-1, 1)
    model_yields = [model.get_yield(tau, X0) for tau in tenors]
    return np.array(model_yields) - actual_yields


def plot_yield_curves(tenors, results):
    """Generates a detailed plot comparing the yield curves across all scenarios."""
    plt.figure(figsize=(12, 7))

    styles = {
        'Baseline': ('r-', 'o', 'darkred'),
        'High_Inflation': ('b--', 's', 'darkblue'),
        'Slow_Growth': ('g-.', '^', 'darkgreen'),
        'Global_Taper': ('m:', 'D', 'darkmagenta')
    }

    for name, yields in results.items():
        line_style, marker, color = styles.get(name, ('k-', 'x', 'black'))
        yields_percent = np.array(yields) * 100

        plt.plot(
            tenors,
            yields_percent,
            line_style,
            marker=marker,
            label=name.replace('_', ' '),
            color=color,
            linewidth=2,
            markersize=7
        )

    plt.title('Projected Zero-Coupon Yield Curves Across RBI Economic Scenarios (2025)', fontsize=16)
    plt.xlabel('Tenor (Years)', fontsize=12)
    plt.ylabel('Yield (%)', fontsize=12)
    plt.legend(title="Economic Scenario", loc='lower right', frameon=True, shadow=True, fancybox=True)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.xticks(tenors)
    plt.ylim(min(MOCK_YIELD_DATA['yields']) * 100 * 0.9, max(MOCK_YIELD_DATA['yields']) * 100 * 1.2) # Dynamic y-axis
    plt.tight_layout()
    plt.show()

def calibrate_and_simulate(scenarios_engine, yield_data):
    """Core function to calibrate the baseline model and simulate scenarios."""
    tenors = yield_data['tenors']
    actual_yields = yield_data['yields']

    # 1. Calibrate Baseline Model: Find the initial factor state X0
    print("\n--- Calibration (Baseline Scenario) ---")
    base_model = scenarios_engine.adjust_model_params('Baseline')

    # Initial guess for X0 is the P-measure long-run mean
    x0_guess = base_model.theta_phys.flatten()

    # Use Least Squares to find the X0 that best fits the current market yields
    res = least_squares(
        objective_function,
        x0_guess,
        args=(base_model, tenors, actual_yields),
        method='lm',
        verbose=0
    )

    X0_calibrated = res.x.reshape(-1, 1)
    base_model.X0 = X0_calibrated

    print(f"Calibration successful: Calibrated Initial Factors (X0) = {X0_calibrated.flatten()}")
    print(f"Initial Yield Curve Fit Error (RMSE): {np.sqrt(np.mean(res.fun**2)) * 100:.3f} bps")

    # 2. Simulate Scenarios
    results = {}
    scenarios = ['Baseline', 'High_Inflation', 'Slow_Growth', 'Global_Taper']

    for name in scenarios:
        scenario_model = scenarios_engine.adjust_model_params(name)
        # ALL scenarios start from the same calibrated market factor state X0
        scenario_model.X0 = X0_calibrated

        # Calculate the new yield curve based on the scenario's changed parameters (theta, sigma, lambda_risk)
        scenario_yields = [scenario_model.get_yield(tau, scenario_model.X0) for tau in tenors]

        results[name] = scenario_yields
        print(f"Scenario '{name}' loaded.")

    return results, base_model

# =============================================================================
# MAIN EXECUTION
# =============================================================================

# Global variables (Mocking the canvas environment variables)
__app_id = 'affine_model_app'

print(f"Initializing Affine Term Structure Model for Indian Economy (2025-10-30)")
print(f"RBI Repo Rate: {REPO_RATE_RBI * 100}%, Inflation Target: {INFLATION_TARGET * 100}%, GDP Growth: {GDP_GROWTH * 100}%")
print("-" * 75)

try:
    econ_scen = EconomicScenarioEngine()

    # Execute calibration, simulation, and results generation
    results, base_model = calibrate_and_simulate(econ_scen, MOCK_YIELD_DATA)

    # --- Display Results Table ---
    print("\n--- Scenario Analysis Summary (Yield in %) ---")
    print(f"{'Tenor (Y)':<10} | {'Baseline':<12} | {'High Infl':<12} | {'Slow Growth':<12} | {'Global Taper':<12}")
    print("-" * 75)

    for i, t in enumerate(MOCK_YIELD_DATA['tenors']):
        # Convert yields to percentage for display
        baseline_yield = results['Baseline'][i] * 100
        high_infl_yield = results['High_Inflation'][i] * 100
        slow_growth_yield = results['Slow_Growth'][i] * 100
        global_taper_yield = results['Global_Taper'][i] * 100

        print(f"{t:<10.1f} | {baseline_yield:<12.3f} | {high_infl_yield:<12.3f} | {slow_growth_yield:<12.3f} | {global_taper_yield:<12.3f}")

    # --- Generate Chart ---
    plot_yield_curves(MOCK_YIELD_DATA['tenors'], results)

except Exception as e:
    print(f"\nAn error occurred during execution: {e}")
```
